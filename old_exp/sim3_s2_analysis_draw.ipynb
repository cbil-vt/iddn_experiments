## Compare iDDN, JGL, and iDINGO, node 2, 90470

```python
exp_name = "sim3_ggm_three_layer_batch_90470"
comm_iddn, diff_iddn, _, _ = sim3.read_one_res(exp_name, n_sample=200, msk_level=2, mrna_only=False, method='iddn')
comm_jgl, diff_jgl, _, _ = sim3.read_one_res(exp_name, n_sample=200, msk_level=2, mrna_only=False, method='jgl')
comm_idingo, diff_idingo, _, _ = sim3.read_one_res(exp_name, n_sample=200, msk_level=2, mrna_only=False, method='idingo')
```

```python
xrange = np.arange(0.02, 0.81, 0.02)
fig = sim3.draw_F1_pROC_curves_for_methods(
    [comm_iddn, comm_jgl, comm_idingo],
    [diff_iddn, diff_jgl, diff_idingo],
    xlim=[-0.0005, 0.01],
    # xlim=[-0.01, 0.21],
    xrange=xrange,
)
fig.savefig("./temp/node_2_compare.png")
# fig.savefig("./temp/node_2_compare.svg")
```

## mRNA vs. mRNA + others, iDDN without constraints, benefits of multi-omics, 200 + 200, 444445

```python
exp_name = "sim3_ggm_three_layer_batch_444445"
reload((sim3))
n_sample = 200
comm_iddn0, diff_iddn0, _, _ = sim3.read_one_res(exp_name, n_sample=n_sample, msk_level=0, mrna_only=True, method='iddn', subset='mrna')
comm_iddn1, diff_iddn1, _, _ = sim3.read_one_res(exp_name, n_sample=n_sample, msk_level=0, mrna_only=True, method='iddn', subset='mrna_tf')
comm_iddn2, diff_iddn2, _, _ = sim3.read_one_res(exp_name, n_sample=n_sample, msk_level=0, mrna_only=True, method='iddn', subset='mrna_mirna')
comm_iddn3, diff_iddn3, _, _ = sim3.read_one_res(exp_name, n_sample=n_sample, msk_level=0, mrna_only=True, method='iddn', subset='mrna_tf_mirna')
```

```python
xrange = np.arange(0.02, 0.81, 0.02)
fig = sim3.draw_F1_pROC_curves_for_methods(
    [comm_iddn0, comm_iddn1, comm_iddn2, comm_iddn3],
    [diff_iddn0, diff_iddn1, diff_iddn2, diff_iddn3],
    xlim=[-0.003, 0.04],
    xrange=xrange,
)
fig.savefig(f"./temp/node_5_iddn_subsets_n_{n_sample}.png")
```
